<p style="font-size:16px;" >Minimum degree, maximum degree, average degree using the in-built and using: $$<k> = \frac{2L}{N}$$ where $<k>$ is the average degree. We also determine the standart deviation.</p>

```python
deg_dic = AIMS_22_23.degree()
deg_key = list(dict(deg_dic).keys())
deg_values = list(dict(deg_dic).values())
min(deg_values),max(deg_values),np.array(deg_values).mean(), (len(AIMS_22_23.edges)*2)/60,np.array(deg_values).std()
```

```python
sum(np.array(deg_values))
```

<p style="font-size:16px;" >Below we are ploting the degree distribution, the cummulative degree  distribution and the histogram of degree distribution of our network</p>

```python
deg_values = [i/274 for i in deg_values]
plt.figure(figsize=(20,8))
plt.plot(deg_key,deg_values)
plt.xticks(rotation=45)
plt.show()
```

```python
def cumul(l):
    l1=[]
    j = 0
    n = len(l)
    for i in range(n):
        l1.append(j)
        j = j+l[i]
    return l1
```

```python
deg_cum = cumul(deg_values)
deg_cum
plt.figure(figsize=(20,8))
plt.plot(deg_key,deg_cum)
plt.xticks(rotation=45)
plt.show()
```

```python
deg_values = list(dict(deg_dic).values())
plt.figure(figsize=(20,8))
plt.xticks(rotation=45)
plt.hist(deg_values)
plt.show()
```
